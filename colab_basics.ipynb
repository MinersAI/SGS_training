# Google Colab Basics

This short notebook helps explain the basics of using Python notebooks in Google Colab! Its aim is to show what code cells are, what markdown cells are, how to run cells, and why running cells in order is important!

----------

Code cells are the basic building blocks of Python notebooks. They allow users to write python code, then execute it, and view its outputs! 

Below is an example of a very basic Python script that should output the phrase "Hello World!"

```python
print("Hello World!")
```

Markdown cells are the other basic building block of Python notebooks - they allow the user to place plain text in alongside their code. Markdown cells are most often used to help walk users through a notebook (like how these markdown cells are used!)

Each cell is run individually, and are most often run in sequential order. Each cell builds upon the next, storing new variables, defining new functions, or other important tasks! If cells that are meant to be run sequentially are run out of order, the notebook may output an error! In the example below, try running Cell 2 before Cell 1 - Cell 2 should output an error message indicating that the variable "message" has not yet been defined. To resolve this error, first run Cell 1, then Cell 2 and see what happens!

```python
# Cell 1

message = "Hello World!"
```

```python
# Cell 2

print(message)
```

The last thing that you need to understand is how to edit variables. Variables store information, such as numbers, messages (strings), data files, and many other "objects." We can edit the value that a variable stores to change the output of functions that use it later! Try changing the value below to a different number, perhaps from `1` to `10`, and see what happens!

```python
# Cell 1 - try changing the value of number from 1 to any other integer

number = 1
```

```python
# Cell 2 - see how the output changes (and remember to run cell 1 first!)

print(number)
```
